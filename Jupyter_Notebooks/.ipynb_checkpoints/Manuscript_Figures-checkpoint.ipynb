```python
def heterozygosity_scatterplot(snpTable, indiv, pp):
    """Input: snpTable.tsv file (output from multiVCF analyzer), sample name (string)
    Usage: Plot multiallelic SNPs by genomic position.
    """
    fig = plt.figure()
    
    #Read data from snpTable into numpy array
    my_data = genfromtxt(snpTable, delimiter='\t',  dtype=str)
    
    #Subset array, including only SNP values in column matching 'indiv' input
    bool_array = np.where(np.ndarray.flatten(my_data[[0],:] == indiv))[0]
    array_sub = np.ndarray.flatten(my_data[1:, bool_array])
    
    #Iterate over subsetted array, checking whether each SNP matches regexp for Heterozygous calls.
    #Appending heterozygous allele frequencies and genomic positions to list
    idx_list = []
    item_list = []
    for index, item in enumerate(array_sub):
        matchObj2 = re.match( r'([RYWSMK]) \((\d+\.?\d+)\)', item)
        if matchObj2:
            idx_list.append(int(my_data[index + 1, [0]]))
            item_list.append(float(matchObj2.group(2)))

    #Plot scatterplot
    plt.xlabel('Genomic Position', fontsize = 10)
    plt.ylabel('Proportion of Reads with Derived Allele', fontsize = 10)
    title = str(indiv)
    plt.title(title, fontsize = 12)
    plt.scatter(idx_list, item_list, c = "black", s= 5)
    plt.savefig(pp ,format = 'pdf')
    
def heterozygosity_histogram(snpTable, indiv, pp):
    """Input: snpTable.tsv file (output from multiVCF analyzer), sample name (string)
    Usage: Plot proportion of reads containing derived allele for each SNP called as heterozygous by multivcf analyzer
    """
    fig = plt.figure()
    
    #Read data from snpTable into numpy array
    my_data = genfromtxt(snpTable, delimiter='\t',  dtype=str)

    #Subset array, including only SNP values in column matching 'indiv' input
    array_sub = np.ndarray.flatten(my_data[1:, np.where(np.ndarray.flatten(my_data[[0],:] == indiv))[0]])

    #Iterate over subsetted array, checking whether each SNP matches regexp for Heterozygous calls.
    #Appending heterozygous allele frequencies to a list
    item_list = []
    for index, item in enumerate(array_sub):
        matchObj2 = re.match( r'([RYWSMK]) \((\d+\.?\d+)\)', item)
        if matchObj2:
            item_list.append(float(matchObj2.group(2)))

    #Plot histogram
    fig = plt.hist(item_list, bins=range(10,92,2))
    plt.xlabel('Proportion of Reads with Derived Allele', fontsize = 10)
    plt.ylabel('Number of Heterozygous SNPs', fontsize = 10)
    title = "S. mutans Strain Heterozygosity: " + str(indiv)
    plt.title(title, fontsize = 12)
    plt.savefig(pp ,format = 'pdf')

def heterozygosity_histogram_w_homozygotes(snpTable, indiv, pp):
    """Input: snpTable.tsv file (output from multiVCF analyzer), sample name (string)
    Usage: Plot proportion of reads containing derived allele for each heterozygous and homozygous derived SNP
    """
    fig = plt.figure()
    
    #Read data from snpTable into numpy array
    my_data = genfromtxt(snpTable, delimiter='\t',  dtype=str)

    #Subset array, including only SNP values in column matching 'indiv' input
    array_sub = np.ndarray.flatten(my_data[1:, np.where(np.ndarray.flatten(my_data[[0],:] == indiv))[0]])

    #Iterate over subsetted array, checking whether each SNP matches regexp for Homozygous or Heterozygous calls.
    #Appending homozygous/heterozygous allele frequencies to a list
    item_list = []
    for index, item in enumerate(array_sub):
        #Heterozygous calls
        matchObj2 = re.match( r'([RYWSMK]) \((\d+\.?\d+)\)', item)
        #Homozygous calls
        matchObj1 = re.match( r'([AGTC]) \((\d+\.?\d+)\)', item)
        if matchObj2:
            item_list.append(float(matchObj2.group(2)))
        elif matchObj1:
            item_list.append(float(matchObj1.group(2)))
    
    #Plot histogram
    fig = plt.hist(item_list, bins=range(10,102,2))
    plt.xlabel('Proportion of Reads with Derived Allele', fontsize = 10)
    plt.ylabel('Number of SNPs', fontsize = 10)
    title = "S. mutans Strain Heterozygosity: " + str(indiv)
    plt.title(title, fontsize = 12)
    plt.savefig(pp ,format = 'pdf')
    
def heterozygosity_histogram_overlay(indiv, pp, **kwargs):
    """Input: Sample name (string), Multiple snpTable.tsv files (output from multiVCF analyzer)
    Usage: Compare patterns of heterozygous SNP calling using different parameters for the same individual.
    For each experiment, plots proportion of reads containing derived allele for each SNP called as heterozygous by multivcf analyzer
    """
    fig = plt.figure()
    data_dict ={}
    colors = ['red', 'blue', 'green', 'yellow', 'orange', 'purple']
    
    #Iterate over multiple snpTable.tsv files, as specified in **kwargs
    for key, value in kwargs.items():
        
        #Read data from snpTable into numpy array
        my_data = genfromtxt(value, delimiter='\t',  dtype=str)

        #Subset array, including only SNP values in column matching 'indiv' input
        array_sub = np.ndarray.flatten(my_data[1:, np.where(np.ndarray.flatten(my_data[[0],:] == indiv))[0]])
        
        #Iterate over subsetted array, checking whether each SNP matches regexp for Heterozygous calls.
        #Appending heterozygous allele frequencies to a list
        item_list = []
        for index, item in enumerate(array_sub):
            matchObj2 = re.match( r'([RYWSMK]) \((\d+\.?\d+)\)', item)
            if matchObj2:
                item_list.append(float(matchObj2.group(2)))

        #Append allele frequencies to dictionary
        data_dict[key] = item_list

    #Iterate over allele frequencies for each experiment, plot histograms separately
    for index, item in enumerate(data_dict.keys()):
        plt.hist(data_dict[item], alpha=0.2, label=item, color = colors[index], bins=range(10,92,2))
    plt.xlabel('Proportion of Reads with Derived Allele', fontsize = 10)
    plt.ylabel('Number of Heterozygous SNPs', fontsize = 10)
    title = "S. mutans Strain Heterozygosity: " + str(indiv)
    plt.title(title, fontsize = 12)
    plt.legend(loc='upper left')
    plt.savefig(pp ,format = 'pdf')
```

```python
def plot_annotations(gb_record, ax, translator): 
    index_to_keep = []
    for index, feature in enumerate(gb_record.features):
        if feature.type in ['gene', 'source']:
            gb_record.features.remove(feature)
    graph_record = translator().translate_record(gb_record)
    graph_record.plot(ax=ax, strand_in_label_threshold=7, with_ruler=False)
    return ax

def read_bed_cov(bed_dir): 
    # Read coverage data from each bedfile into associated dataframe
    columns = []
    dfs = []
    infiles = !find $bed_dir -name "*bed" 
    for file in infiles:
        with open(file, 'r') as f: 
            sample = '.'.join(file.split('/')[-1].split('.')[0:2])
            df = pd.read_table(file, index_col= 2, sep='\s+', names = ['File', sample])
            df.drop(columns = ['File'], inplace = True)
        dfs.append(df)
    merged = pd.concat(dfs, axis=1).transpose()
    return merged



class MyCustomTranslator_mito(BiopythonTranslator):
    """Custom translator implementing the following theme:
    NOTE: Modified from https://edinburgh-genome-foundry.github.io/DnaFeaturesViewer/examples/custom_biopython_translator.html
    """

    def compute_feature_color(self, feature):
        if feature.type == "tRNA":
            return "yellow"
        if feature.type == "CDS": 
            return "green"
        if feature.type == "rRNA": 
            return "purple"
        if feature.type == "assembly_gap": 
            return "red"
        else:
            return "black"

    def compute_feature_label(self, feature):
        if feature.type == 'rRNA':
            return None
        else:
            return BiopythonTranslator.compute_feature_label(self, feature)

    def compute_filtered_features(self, features):
        """Do not display promoters. Just because."""
        return [
            feature for feature in features
            if (feature.type != "source")
        ]
    
    def compute_feature_fontdict(self, features): 
        return {'font': 'Arial', 'size'   : 6}
```

```python
out = '../Data/FAR004.bed'

cov_df = pd.read_csv(out, sep = '\t', names = \
                     ['Chromosome', 'Start', 'Stop', 'Mean'], \
                     index_col= 'Start')
mid = [start + (stop - start)/2 for stop, start in zip(list(cov_df.Stop), list(cov_df.index))]
cov_df['Mid'] = mid
```

```python
# Path to the GenBank file
file_path = '/mnt/archgen/mutans/99-references/Smutans_UA159/GCA_000007465.2_ASM746v2_genomic.gbff'

# Read the GenBank file
record = SeqIO.read(file_path, "genbank")

# Define start and end positions (1-based indexing)
start_position = 1787000
end_position = 1798000

# Adjust to 0-based indexing for Python
start_index = start_position - 1  # Converts 1-based start position to 0-based
end_index = end_position          # End position is directly usable for slicing

# Extract the region sequence
region_sequence = record.seq[start_index:end_index]

# Filter features that are within the region of interest
region_features = []
for feature in record.features:
    # Check if the feature overlaps with the region of interest
    if feature.location.end >= start_position and feature.location.start <= end_position:
        # Adjust feature location to match the new region
        new_start = max(0, feature.location.start - start_index)
        new_end = min(len(region_sequence), feature.location.end - start_index)
        new_location = FeatureLocation(new_start, new_end, strand=feature.location.strand)
        new_feature = SeqFeature(
            location=new_location,
            type=feature.type,
            qualifiers=feature.qualifiers
        )
        region_features.append(new_feature)

# Create a new SeqRecord for the region
region_record = SeqRecord(
    Seq(region_sequence),
    id=record.id,
    name=record.name,
    description=f"Region from {start_position} to {end_position}",
    features=region_features
)

# Add annotations if desired
region_record.annotations = record.annotations



cm = 1/2.54
reference = '/mnt/archgen/mutans/99-references/Smutans_UA159/GCA_000007465.2_ASM746v2_genomic.gbff'

#print (seg)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18*cm, 9*cm), sharex=False, gridspec_kw={'height_ratios': [2,1]}, constrained_layout=True)


ax2.fill(list(cov_df.Mid), list(cov_df.Mean),
           color='blue',       # The outline color
           alpha=0.2)
ax2.set_xlim(1787000, 1798000)
plot_annotations(region_record, ax=ax1, translator=MyCustomTranslator_mito)
plt.savefig("./FAR004_preprint.pdf",  dpi=300)
```
